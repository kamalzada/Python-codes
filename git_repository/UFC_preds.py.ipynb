```python
SOME_NUMERIC_FEATURES = [f for f in df.columns if df[f].dtype=='int64' or df[f].dtype=='float64'
                        and f!='winner' and f!='better_rank']
print(f'The total number of rows that contain outliers before: {df.loc[(abs(zscore(df[SOME_NUMERIC_FEATURES])) > 3).any(axis = 1)].shape[0]}')
```

```python
# Coercing outliers via IQR

def impute_outliers(value):
    if value > upperlimit:
        value = upperlimit
    elif value < lowerlimit:
        value = lowerlimit
    return value

for feature in SOME_NUMERIC_FEATURES:
    Q3 = df[feature].quantile(q = 0.75)
    Q1 = df[feature].quantile(q = 0.25)
    IQR = Q3 - Q1
    outlier_range = IQR * 2.2
    upperlimit = Q3 + outlier_range
    lowerlimit = Q1 - outlier_range
    df[feature] = df[feature].apply(impute_outliers)
    
while df.loc[(abs(zscore(df[SOME_NUMERIC_FEATURES])) > 3).any(axis = 1)].shape[0] != 0:
    df.drop(index = df.loc[(abs(zscore(df[SOME_NUMERIC_FEATURES])) > 3).any(axis = 1)].index, inplace = True)    
print(f'The total number of rows that contain outliers now: {df.loc[(abs(zscore(df[SOME_NUMERIC_FEATURES])) > 3).any(axis = 1)].shape[0]}')
```

```python
'''Coercing outliers via 
mean+3td method'''
def coerce_outliers(col_name): 
    age_values = df[col_name].unique() 
    upper_border = df[col_name].mean() + 3*(df[col_name].std())
    lower_border = df[col_name].mean() - 3*(df[col_name].std()) 
    for num in age_values:
        if num>upper_border:
            df[col_name].replace({num: upper_border}, inplace=True) 
        elif num<lower_border:
            df[col_name].replace({num: lower_border}, inplace=True) 
    return df[col_name]  
```

```python
for f in df.columns:
    if f!='winner' and f!='better_rank':
        coerce_outliers(f)
```

Modelling: Logistic Regression (LR) and Random Forest (RF)

```python
X = df.drop('bet_correct', axis=1)
y = df['bet_correct']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.10,
                                                   stratify=y)
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
oversampling = ADASYN(sampling_strategy = 0.95, random_state = 42)
X_resampled, Y_resampled = oversampling.fit_resample(X_train, y_train)
```

```python
log_reg = LogisticRegression(max_iter=1000)
log_model = log_reg.fit(X_resampled, Y_resampled) 
```

```python
y_pred = log_reg.predict(X_test)
```

```python
log_reg.score(X_resampled, Y_resampled)
```

```python
log_reg.score(X_test,y_test)
```

```python
print(classification_report(y_test, y_pred))
```
